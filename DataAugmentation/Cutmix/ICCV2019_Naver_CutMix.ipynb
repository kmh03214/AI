# Dog, Cat 데이터 준비

## 데이터는 따로 다운 받아서 경로 설정을 해줘야 합니다.
## 해당 글은 ICCV 2019에서 Naver연구팀이 발표한 CutMix를 구현해 봄에 초점을 맞췄습니다.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

os.listdir()

# loading img

import PIL.Image as Image

def resize_img(img,size):
    return img.resize(size)

def load_img(file_path, size):
    data = []
    print(file_path + '/')
    for f in os.listdir(file_path):
        data.append(resize_img(Image.open(file_path + '/' + f) , size))
    return data

img_size = (150,150)

train_cats = load_img('../../../data/my_cat_dog/train/cat', img_size) # img_data_list / element is image not nparray
test_cats = load_img('../../../data/my_cat_dog/test/cat', img_size)
val_cats = load_img('../../../data/my_cat_dog/val/cat', img_size)

train_dogs = load_img('../../../data/my_cat_dog/train/dog', img_size)
test_dogs = load_img('../../../data/my_cat_dog/test/dog', img_size)
val_dogs = load_img('../../../data/my_cat_dog/val/dog', img_size)


Image.open('../../../data/my_cat_dog/train/cat/cat.1.jpg')

#len(train_cats),len(test_cats),len(val_cats),len(train_dogs),len(test_dogs),len(val_dogs)
```

# We have to img -> array & labeling solution

```python
def img_to_array(img):
    return np.array(img, dtype = 'float32')/255.0

train_dogs_arr,train_dogs_sol = np.array([img_to_array(dog) for dog in train_dogs]),np.array([(0,1)]*len(train_dogs))
test_dogs_arr,test_dogs_sol = np.array([img_to_array(dog) for dog in test_dogs]),np.array([(0,1)]*len(test_dogs))
val_dogs_arr,val_dogs_sol = np.array([img_to_array(dog) for dog in val_dogs]),np.array([(0,1)]*len(val_dogs))

train_cats_arr,train_cats_sol = np.array([img_to_array(cat) for cat in train_cats]),np.array([(1,0)]*len(train_cats))
test_cats_arr,test_cats_sol = np.array([img_to_array(cat) for cat in test_cats]),np.array([(1,0)]*len(test_cats))
val_cats_arr,val_cats_sol = np.array([img_to_array(cat) for cat in val_cats]),np.array([(1,0)]*len(val_cats))

train_img,train_sol = np.concatenate((train_dogs_arr,train_cats_arr)),np.concatenate((train_dogs_sol,train_cats_sol))
test_img,test_sol = np.concatenate((test_dogs_arr,test_cats_arr)),np.concatenate((test_dogs_sol,test_cats_sol))
val_img,val_sol = np.concatenate((val_dogs_arr,val_cats_arr)),np.concatenate((val_dogs_sol,val_cats_sol))
```

# CutMix 구현 Naver ICCV 2019

```python
import random
def rand_bbox(size, lam):
    
    W = size[0]
    H = size[1]
    cut_rat = np.sqrt(1. - lam)
    cut_w = np.int(W * cut_rat)
    cut_h = np.int(H * cut_rat)

    # uniform
    cx = np.random.randint(W)
    cy = np.random.randint(H)

    bbx1 = np.clip(cx - cut_w // 2, 0, W)
    bby1 = np.clip(cy - cut_h // 2, 0, H)
    bbx2 = np.clip(cx + cut_w // 2, 0, W)
    bby2 = np.clip(cy + cut_h // 2, 0, H)

    ratio = abs(bbx1-bbx2)*abs(bby1-bby2) / (W*H)

    return bbx1, bby1, bbx2, bby2, ratio

gen_img,gen_sol = [], []

def cut_generator(img_datas,target_datas, size):

    rand_ind = int(random.uniform(0,len(img_datas)))
    rand_ind2 = int(random.uniform(0,len(img_datas)))

    img1,target1 = img_datas[rand_ind].copy(), target_datas[rand_ind].copy()
    img2,target2 = img_datas[rand_ind2], target_datas[rand_ind2]
    
    bbx1,bby1,bbx2,bby2,ratio = rand_bbox(size ,np.random.beta(1,1)) # beta(alpha,alpha) alpha = 1

    bbx1,bbx2,bby1,bby2 = min(bbx1,bbx2),max(bbx1,bbx2),min(bby1,bby2),max(bby1,bby2)

    target1 = target1*(1-ratio) + target2*(ratio)
    img1_o = img1.copy() # 원본
    img1[bbx1:bbx2,bby1:bby2,:] = img2[bbx1:bbx2,bby1:bby2,:]
    img1_c = img1.copy()
    target1_c = target1.copy()

    gen_img.append(img1_c)
    gen_sol.append(target1_c)
    img1 = img1_o.copy()

for i in range(10000):
    cut_generator(train_img, train_sol, img_size)
```
