```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
def target_func(x):
    return 3*x + 5 # w0 = 3 , w1 = 5
```

```python
x = np.array([random.randint(-20,50) for i in range(50)]) # np.linspace(-20,50,50)
y = target_func(x)
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
t = y + 50 * np.random.rand(len(x))
plt.plot(x,t,'o')
plt.grid()
plt.show()
```

```python
def loss(y,y_):
    return np.mean((y-y_)**2)

def gradient(x,y,lr,w0,w1):    
    w0_ = lr*np.mean((y-(w0*x+w1))*x)
    w1_ = lr*np.mean(y-(w0*x+w1))
    return np.array([w0_,w1_])

def gradient_desc(lr,iterate):
    w0,w1 = (3,14)
    for i in range(iterate):
        w0,w1 = lr*gradient(x,y,lr,w0,w1)
        l = loss(y,w0*x+w1)
        print(w0,w1)
        if l < 0.01:
            break
    return w0,w1
    
    
w0,w1 = gradient_desc(0.00586618,10000)
```

```python
pred = w0*x+w1
```

```python
plt.plot(x,t,'o')
plt.grid()

plt.plot(x,pred,'o')
plt.grid()
plt.show()
```
