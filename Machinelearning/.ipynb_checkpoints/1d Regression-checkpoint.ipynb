```python
x,y = np.array(df.weight), np.array(df.height)
```

```python
len(x),len(y)
```

```python
import numpy as np

def calcGrad(x,t,w0,w1):
    y = w0*x + w1
    d_w0 = 2 * np.mean((y-t)*x)
    d_w1 = 2 * np.mean(y-t)
    return d_w0, d_w1
    
def GradDsctOpt(x,t):
    w_init = [10.0,15.0] # 초기 매개변수 
    alpha = 0.00005 # 학습률
    i_max = 1000000 # 최대 반복횟수
    eps = 0.1 # 반복을 종료할 기울기값 한계
    w0 = w_init[0]
    w1 = w_init[1]
    for i in range(1, i_max):
        dmse = calcGrad(x,t,w0,w1)
        w0 = w0 - alpha * dmse[0]
        w1 = w1 - alpha * dmse[1]
        if i % 10000 == 0:
        if max(np.absolute(dmse)) < eps:
            break
            
    return w0,w1

W0,W1 = GradDsctOpt(x,y)
```

```python
print('W=[{0:.6f}, {1: .6f}]'.format(W0,W1))
```

```python
import matplotlib.pyplot as plt

def mse_line(x,t,w):
    y = w[0] * x + w[1]
    mse = np.mean((y-t)**2)
    return mse

def show_line(w):
    xb = np.linspace(40, 90, 100)
    y = w[0] * xb + w[1]
    plt.plot(xb,y,color=(.5,.5,.5), linewidth =4)

plt.figure(figsize=(4,4))
W = np.array([W0,W1])
mse = mse_line(df.weight,df.height,W)
print("w0={0:.3f}, w1={1:.3f}".format(W0,W1))
print("SD = {0:.3f} cm ". format(np.sqrt(mse)))
show_line(W)
plt.plot(x, y, marker='o', linestyle = 'None',
        color = 'cornflowerblue', markeredgecolor = 'black')
plt.xlim(40,90)
plt.grid(True)
plt.show()
```
